```python
#Bai3:
import math
a, b, c = map(int, input("Nhap vao 3 so: ").split())
if a+b>c and b+c>a and a+c>b:
    if a == b ==c:
        print("Tam giac deu")
    elif (a==b)!=c or (a==c)!=b or (b==c)!=a:
        print("Tam giac can")
    elif a**2+b**2==c**2 or a**2+c**2==b**2 or b**2+c**2==a**2:
        print("Tam giac vuong")
    else:
        print("Tam giac binh thuong")
else:
    print("Khong la tam giac")
p=(a+b+c)/2
print("Dien tich", math.sqrt(p*(p-a)*(p-b)*(p-c)))
print("Chu vi: ", 2*p)
```

```python
#Bai5:
import math
a, b, c = map(float, input("Nhap vao 3 so: ").split())
denta=b**2-4*a*c
if denta>0:
    x1=(-b+math.sqrt(denta))/2*a
    x2=(-b-math.sqrt(denta))/2*a
    print(x1, x2)
elif denta==0:
    x1=x2=-b/2*a
    print(x1, x2)
else:
    print("Phuong trinh vo nghiem")
```

```python
#Bai8:
a, b, c, d, m, n = map(int, input("Nhap vao cac gia tri: ").split())
detA=a*d-b*c
if detA==0:
    print("He vo nghiem")
else:
    detA1b=m*d-n*b
    detA2b=a*n-c*m
    x1=detA1b/detA
    x2=detA2b/detA
    print("Hai nghiem cua phuong trinh: ", x1, x2)
```
