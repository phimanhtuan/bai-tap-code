```python
#Bai5
a, b=map(int, input("Nhap vao 2 so a, b: ").split())
if a>b:
    print(a)
elif a<b:
    print(b)
else:
    print("2 so bang nhau")
```

```python
#Bai6
kt=str(input("Nhap vao 1 ki tu: "))
if 'a' <= kt <= 'z' or 'A' <= kt <= 'Z':
    print(kt, "la ki tu chu cai")
elif '0' <= kt <= '9':
    print(kt, "la ki tu chu so")
```

```python
#Bai7
diem=float(input("Nhap vao so diem: "))
if diem>=8:
    print("Hoc luc Gioi")
elif diem>=6.5 and diem<8:
    print("Hoc luc Kha")
elif diem>=5 and diem<6.5:
    print("Hoc luc TB")
elif diem<5:
    print("Hoc luc Yeu")
```

```python
#Bai8
nam=int(input("Nhap vao so nam: "))
if (nam%400==0) or (nam%4==0 and nam%100!=0):
    print(nam, "la nam nhuan")
else:
    print(nam, "khong phai nam nhuan")
```

```python
#Bai9
so=int(input("Nhap vao 1 so: "))
if 0<=so<=9:
    match so:
        case 0:
            print("Khong")
        case 1:
            print("Mot")
        case 2:
            print("Hai")
        case 3:
            print("Khong")
        case 4:
            print("Bon")
        case 5:
            print("Nam")
        case 6:
            print("Sau")
        case 7:
            print("Bay")
        case 8:
            print("Tam")
        case 9:
            print("Chin")
else:
    print("Vui long chay lai code va nhap lai so")
```
